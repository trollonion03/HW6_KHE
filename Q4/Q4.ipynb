```python
import matplotlib.pyplot as plt
import csv

plt.rcParams["font.family"] = "Malgun Gothic"
def main():
    f = open("q4.csv")
    data = csv.reader(f)
    next(data)
    next(data)
    
    sub_data=[[],[],[]]
    sub_data_30=[[],[],[]]

    for row in data:
        for i in range(3, 7):
            row[i] = int(row[i])
                
        on = row[3] + row[5]
        off = row[4] + row[6]
        onoff = on+off
        sub_data[0].append([on, row[2]])
        sub_data[1].append([off,  row[2]])
        sub_data[2].append([onoff, row[2]])

    for i in range(3):
        sub_data[i].sort(reverse=True)
        sub_data[i] = sub_data[i][:30]
   

    title =['get on', 'get off', 'get on+off']
    plt.figure(figsize=(40,10))

    for i in range(3):
        plt.rc('xtick', labelsize=8)
        plt.subplot(3,1,i+1)
        plt.title(title[i])
        plt.ylabel("")
        plt.xticks()
    
        pdata = [[], [], []]
        for ii in range(30):
            pdata[0].append(sub_data[i][ii][0])
            pdata[1].append(sub_data[i][ii][1])
        plt.bar(pdata[1], pdata[0],color='purple')

    plt.show()
    # file close 
    f.close()



if __name__ == "__main__":
    main()
```
